```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.preprocessing import image
import cv2

# Load pre-trained model
model_path = "/content/drive/MyDrive/ColabNotebooks/age_prediction_model.h5"  # Update path
model = load_model(model_path)

# Print model summary to check layer names
model.summary()


layer_name = "conv1_conv"

activation_layer = model.get_layer(layer_name).output


activation_model = Model(inputs=model.input, outputs=activation_layer)

# Load and preprocess an image
img_path = "/content/drive/MyDrive/ColabNotebooks/sample_face.jpg"  # Update with an actual image path
img = image.load_img(img_path, target_size=(224, 224))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
img_array = img_array / 255.0

# Get activations
activations = activation_model.predict(img_array)

# Plot activation maps
fig, axes = plt.subplots(4, 8, figsize=(15, 10))  # Adjust grid size based on filter count
axes = axes.flatten()

for i in range(min(len(activations[0, :, :, :]), len(axes))):
    axes[i].imshow(activations[0, :, :, i], cmap="viridis")
    axes[i].axis("off")

plt.show()
```
